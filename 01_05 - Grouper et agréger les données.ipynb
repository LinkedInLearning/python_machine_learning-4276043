<h2 style="text-align: center; background-color: #4d648d; font-family:Arial; color: white; padding: 13px; line-height: 1; border-radius:10px">Grouper et agréger les données ➕</h2>

```python
import pandas as pd
df = pd.read_csv("db_customers.csv")
df.head()
```

```python
df_grouped = df.groupby(['customer_unique_id', 'customer_state']).agg({"payment_value" : "sum", "order_id" : "count"}).reset_index()
df_grouped.head()
```
