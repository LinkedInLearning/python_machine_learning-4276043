```python
import pandas as pd
df = pd.read_csv("db_customers.csv")
df.head()
```

```python
df2 = pd.read_csv("db_reviews.csv")
df2.head()
```

```python
dataframe_total = pd.merge(df, df2, left_on = "order_id", right_on = "order_id", how = 'outer')
dataframe_total.head()
```
